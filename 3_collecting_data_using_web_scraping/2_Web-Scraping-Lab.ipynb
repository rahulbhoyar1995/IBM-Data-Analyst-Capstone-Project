## Extract information from the given web site
You will extract the data from the below web site: <br>


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
from bs4 import BeautifulSoup
import requests
```

Download the webpage at the url


```python
# get the contents of the webpage in text format and store in a variable called data
data  = requests.get(url).text
```

```python
print("Data from web :")
print("-"*100)
print(data)
```

Create a soup object


```python
soup = BeautifulSoup(data,"html5lib")
```

Scrape the `Language name` and `annual average salary`.


```python
table = soup.find('table')
```

```python
#Get all rows from the table

for row in table.find_all('tr'): # in html table row is represented by the tag <tr>
    # Get all columns in each row.
    cols = row.find_all('td') # in html a column is represented by the tag <td>
    a1 = cols[0].getText()
    a2 = cols[1].getText() # store the value in column 3 as color_name
    a3 = cols[2].getText() # store the value in column 4 as color_code
    a4 = cols[3].getText()
    a5 = cols[4].getText()
    print("{}--{}--{}--{}--{}".format(a1,a2,a3,a4,a5))
```

```python
import pandas as pd

# Assuming you have a list of column names
columns = ["No.", "Language", "Created By", "Average Annual Salary", "Learning Difficulty"]

# Create empty lists for each column
data = {col: [] for col in columns}

# Get all rows from the table
for row in table.find_all('tr')[1:]:  # assuming 'table' is your BeautifulSoup object representing the HTML table
    cols = row.find_all('td')
    if len(cols) == len(columns):  # Ensure that the number of columns in the row matches the expected number
        for col_index, col_name in enumerate(columns):
            data[col_name].append(cols[col_index].getText())

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

Save the scrapped data into a file named *popular-languages.csv*


```python
# Save the DataFrame to a CSV file
df.to_csv("popular-languages.csv", index=False)
```
